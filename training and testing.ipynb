```python
from NERDA.models import NERDA
```

```python
def process_data(data):
     with open(data, 'r',encoding="utf-8") as file:
        
        files = file.readlines()
        a = []
        b = []
        c = []
        d = []
        for i in files:
           
            if i != '\n':
                text = i.split("\t")[0].strip()
                a.append(text)
                lag = i.split("\t")[1].strip()
                b.append(lag) 

            else:
                c.append(a)
                d.append(b)
                a = []
                b = []
                continue
        sentence = {'sentences':c,'tags': d}
        
        return sentence
         
```

```python
def list_of_groups(init_list, childern_list_len):
    list_of_group = zip(*(iter(init_list),) *childern_list_len) # zip(childern_list_len ge list_iterator object) 
    end_list = [list(i) for i in list_of_group] # i is a tuple
    count = len(init_list) % childern_list_len
    end_list.append(init_list[-count:]) if count !=0 else end_list
    return end_list
```

```python
initial_training = process_data("train.iob2")
initial_validation = process_data("valid.iob2")
a  = list_of_groups(initial_training['sentences'],17000)
a1 = list_of_groups(initial_training['tags'],17000)
b  = list_of_groups(initial_validation['sentences'],3000)
b1  = list_of_groups(initial_validation['tags'],3000)
training = {'sentences':a[0],'tags': a1[0]}
validation = {'sentences':b[0],'tags': b1[0]}
test = {'sentences':a[1]+b[1],'tags': a1[1] + b1[1]}
```

```python
transformer = '/biobert-base-cased-v1.1/'
dropout = 0.5
# hyperparameters for trainingtraining
hyperparameters = { 'epochs' :10, 'warmup_steps' : 500, 'train_batch_size':50, 'learning_rate': 0.0001}
tag_scheme = ['B-cell_type', 'I-cell_type','B-protein', 'I-protein', 'B-RNA', 'I-DNA', 'B-cell_line', 'I-cell_line', 'B-DNA', 'I-RNA']
```

```python
model = NERDA(
dataset_training = training,
dataset_validation = validation,
tag_scheme = tag_scheme, 
tag_outside = 'O',
transformer = transformer,
dropout = dropout,
hyperparameters = hyperparameters
)
```

```python
model.train()
```

```python
model.evaluate_performance(test)
```
